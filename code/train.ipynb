```python
from data_processors.event_logs import EventLog, get_window_size
log_name, fold = 'env_permit', '0'
event_log = EventLog(log_name, fold, start_of_suffix=True)
window_size = get_window_size('auto', event_log)
log_dict = event_log.to_dict()
```

```python
from encoders_and_decoders import EventTransformer
import config
from training import fit

model = EventTransformer(cat_attributes = log_dict['cat_attributes'], # list of dictionary with attribute info
                 num_attributes = log_dict['num_attributes'], # list of numerical attribute name
                 embedding_size = config.DEFAULT_EMB_SIZE,
                 encoder_layers = config.DEFAULT_ENCODER_BLOCKS,
                 decoder_layers = config.DEFAULT_DECODER_BLOCKS,
                 encoder_attn_heads = config.DEFAULT_ATTENTION_HEADS,
                 )

model = fit(model, event_log, window_size, epochs = 5)
```

```python
from evaluation import test
similarities = test(model, event_log, window_size)
print(similarities)
```
